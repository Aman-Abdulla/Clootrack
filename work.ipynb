```python
import pandas as pd
```

```python
df = pd.read_csv("C://Users//amana//Documents//GitHub Repository/clootrack//Sample Dataset for Coding Task.csv")
```

```python
df
```

```python
# Calculate total sales for each category
category_sales = df.groupby('Category')['Sales'].transform('sum')
```

```python
# Calculate sales percentage for each product within its category
df['Sales Percentage'] = (df['Sales'] / category_sales) * 100
df['Sales Percentage'] = df['Sales Percentage'].round(2)
```

```python
# Save the result to a CSV file
output_file = 'Aman_Abdulla_products.csv'
df.to_csv(output_file, columns=['Category', 'Product', 'Sales Percentage'], index=False)

print("Output saved to:", output_file)
```
